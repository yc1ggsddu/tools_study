```python
N = 20 # 20个区域
theta = np.linspace(0.0, 2*np.pi, N, endpoint=False) # 0-2pi等分出角度
radii = 10 * np.random.rand(N) # 生成每个角度对应的值
width = np.pi / 4 * np.random.rand(N) # 对应宽度
# 以上为生成的数组
```

```python
ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width, bottom=0.0)

for r,bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)
plt.show()
```
