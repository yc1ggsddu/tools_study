```python
import matplotlib.pyplot as plt
labels = ['frogs', 'hogs', 'dogs', 'logs']
size = [1,3,5,23]
explode = (0, 0.1, 0, 0) #突出
```

```python
plt.pie(x=size, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=False, startangle=90) #autopct中间显示百分数的方式
plt.axis("equal")
plt.show()
```
