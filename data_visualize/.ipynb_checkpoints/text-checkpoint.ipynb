```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(0.0, 6.0, 0.02)
plt.plot(a, np.cos(2*np.pi*a), 'r--')
plt.xlabel("横轴：实践", fontproperties="simhei", fontsize=15, color='green')
plt.ylabel("纵轴：振幅", fontproperties='simhei', fontsize=15)
plt.title(r'正弦波实例 $y=cos(2\pi x)$', fontproperties='simhei', fontsize=20)
plt.text(3, 0, r'$\mu=100$', fontsize=15)
plt.annotate
plt.axis([0, 6, -1, 1])
plt.grid(True)
plt.show()
```
