```python
import matplotlib.pyplot as plt
```

```python
plt.plot([3,1,2,3,4]) #输入只有一维，会当作y轴内容，x轴作为索引
plt.ylabel("grade")
plt.show()
```

```python
plt.savefig("test", dpi=600) #dpi指定保存画质
```

```python
plt.plot([0,2,4,6,8], [3,2,4,5,2])
plt.ylabel("Grade")
plt.axis([-1, 9, 0, 7]) #一个带四个元素的列表，对应x轴及y轴的表示范围
plt.show()
```

```python
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

a = np.arange(0.0, 5.0, 0.02)
plt.subplot(2, 1, 1)
plt.plot(a, f(a))

plt.subplot(2, 1, 2)
plt.plot(a, np.cos(2*np.pi*a), "r--")
plt.show()
```
